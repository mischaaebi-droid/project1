```python
import requests
import time
import json
import time
import random
demand_list = []
counter=0
for number in range(3000, 10000):
    url = f"https://ws-old.parlament.ch/affairs/2025{number}"
    counter=counter+1
    print("count: "+str(counter))
    r = requests.get(url, headers={"Accept": "application/json"})

    if r.status_code == 200:
        demand_list.append(r.json())
    else:
        print(f"Nicht gefunden: 2025{number}")

    data = r.json()
    demand_list.append(data)

    with open("geschaefte.jsonl", "a", encoding="utf-8") as f:
        f.write(json.dumps(data, ensure_ascii=False))
        f.write("\n")
        time.sleep(random.uniform(0.3, 1.5))

    
```

```python
import json
import pandas as pd

rows = []

with open("geschaefte.jsonl", encoding="utf-8") as f:
    for line in f:
        rows.append(json.loads(line))

print(len(rows))
```

```python
flat = []

for g in rows:

    if "councillor" not in g.get("author", {}):
        continue

    flat.append({"person_code": g["author"]["councillor"]["code"],"author": g["author"]["councillor"]["name"],"type": g["affairType"]["name"]})

df = pd.DataFrame(flat)
```

```python
df_Int = (
    df[df["type"] == "Interpellation"]
    .groupby(["person_code", "author"])
    .size()
)

df_Post = (
    df[df["type"] == "Postulat"]
    .groupby(["person_code", "author"])
    .size()
)

df_Mot = (
    df[df["type"] == "Motion"]
    .groupby(["person_code", "author"])
    .size()
)

df_tot = (
    df.groupby(["person_code", "author"])
    .size()
)
```

```python
df_stats = pd.DataFrame({
    "total": df_tot,
    "motionen": df_Mot,
    "postulate": df_Post,
    "interpellationen": df_Int
}).fillna(0)
```

```python
df_stats = df_stats.reset_index()
```

```python
import json

liste = df_stats.to_dict(orient="records")

with open("parlamentarier_stats.json", "w", encoding="utf-8") as f:
    json.dump(liste, f, ensure_ascii=False, indent=2)
```
