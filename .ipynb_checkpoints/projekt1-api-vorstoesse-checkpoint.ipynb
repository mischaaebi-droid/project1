```python



url_hotinfo_sanct = "https://api.ebird.org/v2/data/obs/L39789806/recent?back=30"

headers = {
    "X-eBirdApiToken": "r3n4k858o9v4"
}

response_hotinfo_sanct = requests.get(url_hotinfo_sanct, headers=headers)

data_info_sanct=response_hotinfo_sanct.json()

print(response_hotinfo_sanct.status_code)
#pprint.pprint(data_info_sanct)


numb_birds=0
numb_warbler=0
observations=len(data_info_sanct)
print(f"number of obesrvations {observations}")
for birdspecies in data_info_sanct:
    #print (birdspecies['comName'], birdspecies['howMany'])
    numb_birds=numb_birds+birdspecies['howMany']
    if "Warbler" in birdspecies['comName']:
        print (f"That is a {birdspecies['comName']}")
        numb_warbler=numb_warbler+birdspecies['howMany']
        percent=100/numb_birds*numb_warbler
        
print(f"total number of birds: {numb_birds}")
print(f"total number of Warblers: {numb_warbler}")
print(f"Percent of Warblers: {percent}")
    
```
