Here in de following, you can always retrieve all transactions for an entire calendar year, along with all associated data. For multiple years or an entire legislative term, you can run the program multiple times, changing the year each time. The transactions are saved continuously.

In the following, the requested data is read from the JSON file, processed statistically, and evaluated.

```python
import json
import pandas as pd

rows = []

with open("geschaefte.jsonl", encoding="utf-8") as f:
    for line in f:
        rows.append(json.loads(line))

states= rows[2]["drafts"][0]['states']
state_list = []
for state in states:
    state_list.append(state["id"])
if 222 in state_list:
    success=1
    
```

```python
flat = []

for g in rows:

    if "councillor" not in g.get("author", {}):
        continue

    states = g["drafts"][0]["states"]

    state_list = []
    for state in states:
        state_list.append(state["id"])

    if 209 in state_list:
        success = 1
    else:
        success = 0

    flat.append({
        "person_code": g["author"]["councillor"]["code"],
        "author": g["author"]["councillor"]["name"],
        "type": g["affairType"]["name"],
        "success": success
    })

df = pd.DataFrame(flat)
df["success"].value_counts()
```

```python
df_Int = (
    df[df["type"] == "Interpellation"]
    .groupby(["person_code", "author"])
    .size()
)

df_Post = (
    df[df["type"] == "Postulat"]
    .groupby(["person_code", "author"])
    .size()
)

df_Mot = (
    df[df["type"] == "Motion"]
    .groupby(["person_code", "author"])
    .size()
)

df_tot = (
    df.groupby(["person_code", "author"])
    .size()
)

df_success_int = (
    df[df["type"] == "Interpellation"]
    .groupby(["person_code", "author"])["success"]
    .sum()
)

df_success_post = (
    df[df["type"] == "Postulat"]
    .groupby(["person_code", "author"])["success"]
    .sum()
)

df_success_mot = (
    df[df["type"] == "Motion"]
    .groupby(["person_code", "author"])["success"]
    .sum()
)

df_success_total = (
    df.groupby(["person_code", "author"])["success"]
    .sum()
)

df_stats = pd.DataFrame({
    "total": df_tot,

    "motionen": df_Mot,
    "postulate": df_Post,
    "interpellationen": df_Int,

    "success_total": df_success_total,
    "success_motionen": df_success_mot,
    "success_postulate": df_success_post,
    "success_interpellationen": df_success_int
}).fillna(0).astype(int)

df_stats["mot_and_post"] = df_stats["motionen"] + df_stats["postulate"]
df_stats["success_mot_and_post"] = (
    df_stats["success_motionen"] + df_stats["success_postulate"]
)
```

```python
df_stats = df_stats.reset_index()
```
